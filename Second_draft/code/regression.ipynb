```python
import numpy as np
import matplotlib.pyplot as plt

# Example input vectors: x holds time in seconds, y holds <R^2(t)> in micron^2.
# Replace these with your actual data.
x = np.array([0.5, 1.0, 1.5, 2.0, 2.5])
y = np.array([1.2, 2.5, 3.8, 5.1, 6.3])

# Number of data points
N = len(x)

# Calculate beta-hat (slope) using the formula:
# beta_hat = (sum(x * y)) / (sum(x^2))
beta_hat = np.sum(x * y) / np.sum(x**2)

# Calculate residuals and unbiased estimate of residual variance sigma^2
residuals = y - beta_hat * x
sigma_hat_squared = np.sum(residuals**2) / (N - 1)
sigma_hat = np.sqrt(sigma_hat_squared)

# Calculate the standard error for beta:
sigma_beta = sigma_hat / np.sqrt(np.sum(x**2))

# Print computed values
print("Estimated beta (slope):", beta_hat)
print("Unbiased estimate of sigma:", sigma_hat)
print("Standard error of beta:", sigma_beta)

# Create scatterplot and overlay the fit line.
fig, ax = plt.subplots(figsize=(6, 5))

# Scatterplot of data
ax.scatter(x, y, color='blue', label='Data')

# Create a fit-line from 0 to max(x)
x_fit = np.linspace(0, np.max(x), 100)
y_fit = beta_hat * x_fit
ax.plot(x_fit, y_fit, color='red', linewidth=2, label=f'Fit: y = {beta_hat:.2f} x')

# Center the axes at 0
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
# Hide the top and right spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Optionally, add arrows to the axes (if desired)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

# Set axis labels
ax.set_xlabel("seconds", fontsize=12)
ax.set_ylabel(r"$\mu m^2$", fontsize=12)

# Add a legend in the bottom-right and grid
ax.legend(loc='lower right', bbox_to_anchor=(1, 0.1))
ax.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
